```python
# Read data using pandas

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('train.csv')
```

```python
# split the data into survivals and casualties

survivals = data[data['Survived'] == 1]
casualties = data[data['Survived'] == 0]

# plot the survivals and casualties based on fare and age

plt.scatter(survivals['Age'], survivals['Fare'], label='Survivals', alpha=0.7)
plt.scatter(casualties['Age'], casualties['Fare'], label='Casualties', alpha=0.7)

# add labels and legend

plt.title('Scatter Plot Survivals and Casualties based on Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()

plt.show()
```

```python
plt.hist(casualties['Embarked'],  label='Casualties')

plt.title('Casualties by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Number')
plt.xticks([1, 2, 3], ['C', 'Q', 'S'])
plt.legend()
plt.show()
```

```python
plt.hist(survivals['Embarked'],  label='Survivals')

plt.title('Survivals by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Number')
plt.xticks([1, 2, 3], ['C', 'Q', 'S'])
plt.legend()
plt.show()
```
